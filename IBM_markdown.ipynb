### The code in the next cell would simply find a pattern in a text using regex 

````python
import re

name = input("Hello, what is your first and last name?:" )
regex = r"^(\w*) (\w*)$"

result = re.search(regex, name)
print(f"Your first name is {result[1]}")
````
